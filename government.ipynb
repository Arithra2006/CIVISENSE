```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Load and clean
df = pd.read_csv("census_2011_india_formatted.csv")

# Preprocessing
df["Population"] = df["Population"].str.replace(",", "").astype(float)
df["Average Income (INR)"] = df["Average Income (INR)"].str.replace(",", "").astype(float)
df["Poverty Rate (%)"] = df["Poverty Rate (%)"].astype(str).str.replace("%", "").astype(float)
df["Literacy Rate (%)"] = df["Literacy Rate (%)"].astype(str).str.replace("%", "").astype(float)

# Train model to predict poverty based on other fields
X = df[["Population", "Literacy Rate (%)", "Average Income (INR)"]]
y = df["Poverty Rate (%)"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model = LinearRegression()
model.fit(X_train, y_train)

print("R² Score:", model.score(X_test, y_test))
```
