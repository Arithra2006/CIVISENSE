```python
# development_analysis.ipynb

import pandas as pd

# Load datasets
df_1991 = pd.read_csv("census_2011_india_formatted.csv")
df_2001 = pd.read_csv("census_2001_india_formatted.csv")
df_2011 = pd.read_csv("census_1991_india_formatted.csv")

# Ensure columns are in the correct format
for df in [df_1991, df_2001, df_2011]:
    for col in ["Population", "Average Income (INR)"]:
        df[col] = df[col].replace(",", "", regex=True).astype(int)
    for col in ["Poverty Rate (%)", "Literacy Rate (%)"]:
        df[col] = df[col].astype(float)

# Save cleaned data
df_1991.to_csv("Census_1991_cleaned.csv", index=False)
df_2001.to_csv("Census_2001_cleaned.csv", index=False)
df_2011.to_csv("Census_2011_cleaned.csv", index=False)
```
